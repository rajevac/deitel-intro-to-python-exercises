```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random
import seaborn as sns
```

```python
die_one = [random.randrange(1, 7) for i in range(36000)]
die_two = [random.randrange(1, 7) for i in range(36000)]
die_sum = [x + y for (x, y) in zip(die_one, die_two)]
print(min(die_sum))

values, frequencies = np.unique(die_sum, return_counts=True)

sns.set_style('whitegrid')
plt.figure(figsize=(13, 10))
axes = sns.barplot(x=frequencies, y=values, palette='bright', orient='h')
axes.set_title(f'Frequency of Sums Tossing Two Die {len(die_sum):,} Times')
axes.set(xlabel='Frequency', ylabel='Sum of Two Die')
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_width() * 1.025
    text_y = bar.get_y() + bar.get_height() / 2.0
    text = f'{frequency:,}\n{frequency / len(die_sum):.3%}'
    axes.text(text_x, text_y, text, fontsize=11, ha='left', va='center_baseline')
```
