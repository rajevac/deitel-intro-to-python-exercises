```python
import numpy as np
import random
import seaborn as sns

coin_tossing = [random.randrange(1, 3) for i in range(200)]
values, frequencies = np.unique(coin_tossing, return_counts=True)

sns.set_style('whitegrid')
axes = sns.barplot(x=values, y=frequencies, palette='bright')
axes.set_title(f'Tossing a Coin {len(coin_tossing):,} Times')
axes.set(xlabel='Coin Tossing Values: 1–Head, 2–Tail', ylabel='Frequency')
axes.set_ylim(top=max(frequencies)*1.10)
```

```python
coin_tossing = [random.randrange(1, 3) for i in range(200)]
values, frequencies = np.unique(coin_tossing, return_counts=True)

sns.set_style('whitegrid')
axes = sns.barplot(x=values, y=frequencies, palette='bright')
axes.set_title(f'Tossing a Coin {len(coin_tossing):,} Times')
axes.set(xlabel='Coin Tossing Values: 1–Head, 2–Tail', ylabel='Frequency')
axes.set_ylim(top=max(frequencies)*1.20)
for bar, frequency in zip(axes.patches, frequencies):
    text_x = bar.get_x() + bar.get_width() / 2.0
    text_y = bar.get_height()
    text = f'{frequency:,}\n{frequency / len(coin_tossing):.3%}'
    axes.text(text_x, text_y, text, fontsize=11, ha='center', va='bottom')
```
